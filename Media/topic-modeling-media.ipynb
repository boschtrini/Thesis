```python
from umap import UMAP
from hdbscan import HDBSCAN
from sklearn.feature_extraction.text import CountVectorizer
from bertopic import BERTopic
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
media = pd.read_pickle('/kaggle/input/news-data/final-news-data.pkl')
embeddings = pd.read_pickle('/kaggle/input/news-embeddings-bert/embeddings-news-10m.pkl')
```

```python
media['all_text'] = media['Body'].fillna('') +''+media['title'].fillna('')

# Filter rows where 'all_text' is not a string or is empty
# This preserves the DataFrame structure and index
filtered_news = media[media['all_text'].apply(lambda x: isinstance(x, str) and x.strip() != '')]

# Now, if you need just the cleaned texts along with their original indexes:
cleaned_texts_with_indexes = filtered_news['all_text']

# If you need the texts in a list without indexes:
cleaned_texts = cleaned_texts_with_indexes.tolist()

# If you specifically need the indexes:
indexes = cleaned_texts_with_indexes.index.tolist()
```

```python
# Initialize and fit BERTopic
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(cleaned_texts, embeddings)
```

```python
topic_model.get_topic_info().head(50)
```

```python
#Configure UMAP and HDBSCAN models with the specified hyperparameters
umap_model = UMAP(n_neighbors=90, n_components=15, metric='cosine', low_memory=False, random_state=42)
hdbscan_model = HDBSCAN(min_cluster_size=120, metric='euclidean', prediction_data=True, min_samples=5)
    
# Initialize and fit BERTopic
topic_model2 = BERTopic(hdbscan_model=hdbscan_model,
                       umap_model=umap_model,
                       calculate_probabilities=True)
topics2, probs2 = topic_model2.fit_transform(cleaned_texts, embeddings)
```

```python
topic_model2.get_topic_info().head()
```

```python
# Reduce outliers
new_topics = topic_model2.reduce_outliers(cleaned_texts, topics2)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


# Assuming `topic_model2` is your BERTopic model with embeddings calculated
topic_embeddings = topic_model2.topic_embeddings_

# Calculate cosine similarity matrix
cos_sim_matrix = cosine_similarity(topic_embeddings)

# Find topic pairs with high similarity
similarity_threshold = 0.85
similar_topics = []

# Iterate over the upper triangle of the cosine similarity matrix
for i in range(cos_sim_matrix.shape[0]):
    for j in range(i + 1, cos_sim_matrix.shape[1]):
        if cos_sim_matrix[i, j] >= similarity_threshold:
            similar_topics.append((i, j, cos_sim_matrix[i, j]))

# Assuming the BERTopic model has a method to get the top n words for a topic
for i, j, similarity in similar_topics:
    topic_i_name = ', '.join([word for word, _ in topic_model2.get_topic(i)])  # Get the most representative words for topic i
    topic_j_name = ', '.join([word for word, _ in topic_model2.get_topic(j)])  # Get the most representative words for topic j
    similarity_percentage = similarity * 100  # Convert similarity to percentage
    print(f"Topic {i} ({topic_i_name}) and Topic {j} ({topic_j_name}) are {similarity_percentage:.2f}% similar.")
```
